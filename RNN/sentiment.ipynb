```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('IMDB Dataset.csv')
df.head()
```

# text preprocessing

```python
df.drop_duplicates(inplace=True)
```

```python
import re

# Convert all reviews to lowercase
df["review"] = df["review"].str.lower()

# Remove HTML tags
df["review"] = df["review"].apply(lambda x: re.sub(r'<.*?>', '', x))

# Remove special characters and numbers
df["review"] = df["review"].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))




df.head(10)
```

# Removing stopworks using NLTK

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')


stop_words = set(stopwords.words('english'))

df['review'] = df['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
df.head(10)
```

# Stemming and lemm

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer


nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')

# Initialize tools
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()


df['review'] = df['review'].apply(
    lambda x: ' '.join([
        stemmer.stem(word)
        for word in word_tokenize(str(x).lower())
        if word.isalnum() and word not in stop_words
    ])
)


df.head(10)
```

```python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('punkt_tab')  
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')


lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

#Function for preprocessing (clean + reusable)
def preprocess(text):
    tokens = word_tokenize(str(text).lower())
    clean_words = [
        lemmatizer.lemmatize(word)
        for word in tokens
        if word.isalnum() and word not in stop_words
    ]
    return ' '.join(clean_words)

df['review'] = df['review'].apply(preprocess)

df.head(10)
```

# LabelEncoder


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['sentiment'] = le.fit_transform(df['sentiment'])
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer


tf = TfidfVectorizer(max_features=6000)
x = tf.fit_transform(df['review']).toarray()
x.shape
```

# Dataset and Dataloader


```python
from sklearn.model_selection import train_test_split
import torch
from torch.utils.data import DataLoader, TensorDataset, Dataset

x_train, x_test, y_train, y_test = train_test_split(x, df['sentiment'], test_size=0.2, random_state=42)

train_set = TensorDataset(
    torch.tensor(x_train, dtype=torch.float32),
    torch.tensor(y_train.values, dtype=torch.long)
)
test_set = TensorDataset(
    torch.tensor(x_test, dtype=torch.float32),
    torch.tensor(y_test.values, dtype=torch.long)
)

train_dataloader = DataLoader(train_set, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_set, batch_size=64, shuffle=False)
```

# Build RNN Model

```python
import torch.nn as nn
import torch.optim as optim

```

```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(RNN, self).__init__()

        self.hidden_size = hidden_size
        self.num_layers = num_layers

        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, num_classes)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.rnn(x, h0)
        out = self.fc(out[:, -1, :])
        return out
    
```

```python
input_size = x.shape[1]
hidden_size = 128
num_layers = 2
num_classes = 1 

model = RNN(input_size, hidden_size, num_layers, num_classes)


criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)


```

```python
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for Xb, yb in train_dataloader:
        optimizer.zero_grad()

        Xb = Xb.unsqueeze(1)   # (batch, seq_len=1, features)
        yb = yb.float()

        outputs = model(Xb).squeeze(1)  # safer squeeze

        loss = criterion(outputs, yb)

        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}')
```

```python
model.eval()
with torch.no_grad():
    correct, total = 0, 0
    for Xb, yb in test_dataloader:
        Xb = Xb.unsqueeze(1)
        outputs = model(Xb).squeeze(1)
        predicted = (torch.sigmoid(outputs) > 0.5).long()
        total += yb.size(0)
        correct += (predicted == yb).sum().item()

    print(f'Test Accuracy: {correct / total:.4f}')
```
